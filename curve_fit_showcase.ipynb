```python
# Basic imports
import torch
from torch import nn

from assets.shapes import square, circle, stanford_bunny
from assets.loss_functions import mse
from assets.utils import automate_training, plot_curves
```

# Showcase

## Heart

```python
from assets.shapes import heart
from assets.networks import NIG_Net

# Generate target curve points
num_pts = 1000
t = torch.linspace(0, 1, num_pts).reshape(-1, 1)
X_t = heart(num_pts)

nig_net = NIG_Net(
    layer_count_inj = 2, smm_num_grps = 6, smm_neurons_per_grp = 6
)
automate_training(
    model = nig_net, loss_fn = mse, X_train = t, Y_train = X_t,
    learning_rate = 0.1, epochs = 1000, print_cost_every = 200
)

X_p = nig_net(t)
plot_curves(X_p, X_t)
```
